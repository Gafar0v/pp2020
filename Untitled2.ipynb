Загрузка библиотек


```
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.api import ExponentialSmoothing
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_pacf,plot_acf
from statsmodels.stats.outliers_influence import variance_inflation_factor
from pmdarima import auto_arima
%matplotlib inline
```

Подготовка датасета

```
df = yf.download('AAPL','2016-01-02','2020-01-01')
df.drop(['Low', 'High','Open','Adj Close','Volume'], axis='columns', inplace=True)
train=df[:int(0.7*(len(df)))]
test=df[int(0.7*(len(df))):]

```

Подбор коэф для саримы

```
model = auto_arima(train, seasonal=True, m=8, trace=True, suppress_warnings=True, error_action='ignore', stepwise=True)
model
```

Составление модели и построение графиков

```
mod = sm.tsa.statespace.SARIMAX(test,order=(0, 1, 1),seasonal_order=(0, 0, 1, 8))

results = mod.fit()
results.plot_diagnostics(figsize=(30,15))
plt.show()

results.fittedvalues
```

# Построение прогноза

```
predict=results.get_prediction()
ax = train.plot(figsize=(60,60), color='black', title="Прогноз методом SARIMA" )
test.plot(ax=ax, style='--', color='black')
predict.predicted_mean.plot(ax=ax, style='--', color='red')
plt.show()
```
